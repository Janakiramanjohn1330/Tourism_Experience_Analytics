```python
# distribution across continents, countries, and regions.

#set Distribution Across Continents
plt.figure(figsize=(12, 5))
continent_counts = eda['Continent'].value_counts()  # Store value counts
sns.barplot(y=continent_counts.values, x=continent_counts.index, palette='viridis')  
plt.title('User Distribution Across Continents')
plt.xlabel('Continent')
plt.ylabel('Number of user')
plt.show()
```

```python
# User Distribution Across Countries
plt.figure(figsize=(12, 5))
sns.countplot(y=eda['Country'], order=eda['Country'].value_counts().nlargest(10).index,palette='viridis')
plt.title('User Distribution Across Countries')
plt.xlabel('Number of Users')
plt.ylabel('Country')
plt.show()
```

```python
# User Distribution Across Region
plt.figure(figsize=(12, 5))
sns.countplot(y=eda['Region'], order=eda['Region'].value_counts().nlargest(10).index,palette='viridis')
plt.title('User Distribution Across Region')
plt.xlabel('Number of Users')
plt.ylabel('Region')
plt.show()
```

```python
# Attraction Types Popularity Based on Ratings
plt.Figure(figsize=(15,5))
sns.barplot(x=eda['AttractionType'].value_counts().values,y=eda['AttractionType'].value_counts().index)
plt.title('Attraction Types Popularity Based on Ratings')
plt.xlabel('Number of Ratings')
plt.ylabel('Attraction Type')
plt.show()
```

```python
# Correlation Between VisitMode and User Demographics
plt.figure(figsize=(10,5))
sns.countplot(x=eda['VisitMode'],order=eda['VisitMode'].value_counts().index)
plt.xlabel('Visit Mode')
plt.ylabel('Number of Visit')
plt.show()
```

```python

# Bar plot for Average Ratings Across Attractions
plt.figure(figsize=(12, 5))
attraction_avg_ratings = eda.groupby('Attraction')['Rating'].mean().nlargest(10).reset_index()
sns.barplot(x='Rating', y='Attraction', data=attraction_avg_ratings, palette='viridis')
plt.title('Average Ratings Across Attractions')
plt.xlabel('Average Rating')
plt.ylabel('Attraction')
plt.show()
```

```python
# Bar plot for Average Ratings Across Regions
plt.figure(figsize=(12, 5))
region_avg_ratings = eda.groupby('Region')['Rating'].mean().nlargest(10).reset_index()
sns.barplot(x='Rating', y='Region', data=region_avg_ratings, palette='viridis')
plt.title('Average Ratings Across Regions')
plt.xlabel('Average Rating')
plt.ylabel('Region')
plt.show()
```
